```python
f(n) = n*(1+1+1+1) + 2*n*(n-1)*(1+1+1+1) + 5 + 5
plot(f, (n,0,100), axes_labels=['n','f(n)'], 
     title='f(n) = n * (1+1+1+1) + 2 * n * (n-1) * (1+1+1+1) + 5 + 5', 
     title_pos=(0.5,-0.05))
```

```python
f(n) = 10*n^2 + n *10 + 10
g(n) = 100*n^2 + n *100 + 100

f_p = plot(f, (n,0,100))
g_p = plot(g, (n,0,100))
ans = f_p + g_p 
ans += text("g(n)", (80000000, +100000000000000000), fontsize=10, color='black')
ans += text("f(n)", (80000000, +900000000000000000), fontsize=10, color='black')
ans
```

```python
f(n) = 10*n^2 + n *10 + 10
g(n) = 100*n^2 + n *100 + 100
quad(n) = 2*n^2
lin(n) = 1000*n
ex(n) = 2^n

f_p = plot(f, (n,0,50))
g_p = plot(g, (n,0,50))
quad_p = plot(quad, (n,0,50), color='red')
lin_p = plot(lin, (n,0,50),color = 'green')
ex_p = plot(ex, (n,0,50),color = 'brown')

ans = f_p + quad_p + lin_p + ex_p
ans
```
